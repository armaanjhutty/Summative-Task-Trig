```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    equation_type = input("'sin' to plot a sin graph or 'cos' to plot a cos graph or 'tan' to plot a tan graph: ")

    if equation_type.lower() == 'sin':     #https://pythonforundergradengineers.com/plotting-sin-cos-with-matplotlib.html used this website to make sure that the trig functions plot in the intervals of pi, but in my case the range will go up 5*pi in order to see the graph better this was used for all sin and cos
        x = np.linspace(-5 * np.pi, 5 * np.pi, 500)
        y_sin = np.sin(x)
        plt.plot(x, y_sin, label='y = sin(x)')
    elif equation_type.lower() == 'cos':
        x = np.linspace(-5 * np.pi, 5 * np.pi, 500)
        y_cos = np.cos(x)
        plt.plot(x, y_cos, label='y = cos(x)')
    elif equation_type.lower() == 'tan': #https://stackoverflow.com/questions/54505852/how-to-plot-tanx-with-pyplot-and-numpy learned that using the np.linspace would not work with this because it deafaults to 50 elements in the given data set, this results in the graph elements being very widely spaced, instead used code that used both np.linspace in pi interavals with the plt.ylim line to limit the y axis to only 10, allowing for the proper printing of the graph of tan
        x = np.linspace(-5 * np.pi, 5 * np.pi, 1000)
        y_tan = np.tan(x)
        y_tan[:-1][np.diff(y_tan) < 0] = np.nan #https://stackoverflow.com/questions/42837910/omit-joining-lines-in-matplotlib-plot-e-g-y-tanx used this to get this line of code, otherwise the graph was printing with lines in between each tan interval
        plt.plot(x, y_tan)
        plt.ylim(-10, 10)
    else:
        print("equation not allowed only sin,cos,tan graphs")
        return

    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.axhline(0, color='black', lw=1)
    plt.axvline(0, color='black', lw=1)
    plt.show()
main()
```
